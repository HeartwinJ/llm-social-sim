```python
import pandas as pd
```

```python
data = pd.read_csv('summary_final.csv')
data.head()
```

```python
models = ['llama2', 'llama3', 'mistral', 'gemma', 'phi3']
games = ['pd', 'sh', 'hd']
```

```python
new_data = []
for model in models:
	row = []
	row.append(model)
	row.append(data[data['model'] == model]['correct'].sum())
	row.append(data[data['model'] == model]['correct'].sum() / len(data[data['model'] == model]))

	new_data.append(row)

_df = pd.DataFrame(new_data)
_df.columns = ['model', 'correct', 'accuracy']

_df
```

```python
sorted_df = _df.sort_values('accuracy', ascending=False)
sorted_df
```

```python
for game in games:
	new_data = []
	for model in models:
		row = []
		row.append(model)
		row.append(data[(data['model'] == model) & (data['game'] == game)]['correct'].sum())
		row.append(data[(data['model'] == model) & (data['game'] == game)]['correct'].sum() / len(data[(data['model'] == model) & (data['game'] == game)]))

		new_data.append(row)
	
	_df = pd.DataFrame(new_data)
	_df.columns = ['model', 'correct', 'accuracy']
	print(f"Game: {game}")
	display(_df)
```
